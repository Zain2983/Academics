# Scatter Plots


```python
df.plot(kind='scatter',
        x='SepalLengthCm',
        y='SepalWidthCm',
        title='Sepal Lenght vs Sepal Width')
plt.show()
```

```python
ax = sns.scatterplot(x='PetalLengthCm',
                y='PetalWidthCm',
                hue='PetalLengthCm',
                data=df)
ax.set_title('Petal Length vs Petal Width')
plt.show()
```

```python
df.plot(kind='scatter',
        x='SepalWidthCm',
        y='PetalWidthCm',
        title='Sepal Width vs Petal Width')
plt.show()
```

```python
df.plot(kind='scatter',
        x='SepalLengthCm',
        y='PetalLengthCm',
        title='Sepal Length vs Petal Length')
plt.show()
```
